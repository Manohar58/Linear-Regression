Q1 Find the linear regression model which best predicts the dependent variable (“sales”) on validation dataset.

● Divide data in 80-20 % for training and testing set.

● Use 10 fold cross validation set (10% of training dataset).

Find regression co-efficients and intercepts. Report confusion matrix on test dataset.

```python
import pandas as pd
import seaborn as sn
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score 

from sklearn.model_selection import cross_val_score, cross_val_predict
```

```python
DF7 =pd.read_csv("https://raw.githubusercontent.com/manishbhnau/Repo/master/Advertising.csv")
DF7.drop(["Unnamed: 0"],axis = 1, inplace = True);
DF7.head()
```

```python
# printing shape of the DataFrame
DF7.shape
```

```python
# visualize the relationship between the features and the response using scatterplots
sn.pairplot(DF7, x_vars=['TV','radio','newspaper'],y_vars='sales', size=3.5, aspect=0.7)
```

**Televison**

```python
X = DF7[['TV']].values
y = DF7['sales'].values
print(f'Shape of X is {X.shape}\nShape of y is {y.shape}')
```

```python
#Splitting Training and Test Set
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=1)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# fit a model
regression_model = LinearRegression()
model = regression_model.fit(X_train, y_train)
```

```python
print('intercept:', model.intercept_)
print('coefficients for radio:', model.coef_)
```

```python
# Performing 10-fold cross validation
score = cross_val_score(LinearRegression(),X_train,y_train,cv = 10)
print('Cross-validated scores:', score)
```

```python
print('Mean of 10-fold cross validation score : ',score.mean())
```

```python
#Making prediction using model
y_pred = model.predict(X_test)
y_pred
```

```python
# Making cross validated predictions
cross_validate_y_pred = cross_val_predict(model, X_test, y_test, cv=10)
cross_validate_y_pred
```

```python
from sklearn.metrics import mean_squared_error
print('RMS value of test and model predicted value :',mean_squared_error(y_test,y_pred))
print('RMS value of test and cross validate predicted value :',mean_squared_error(y_test,cross_validate_y_pred))
```

```python
import numpy as np
DF8 = pd.read_csv("/auto-mpg.csv")

DF8.head()

DF8.info()

DF8 = DF8.drop('car name', axis=1)

pd.unique(DF8.origin)

DF8['origin'] = DF8['origin'].replace({1: 'america', 2: 'europe', 3: 'asia'})


DF8 = pd.get_dummies(DF8, columns=['origin'], drop_first=True)

DF8.head()

DF8.describe()

DF8 = DF8.replace('?', np.nan)

DF8[DF8.isnull().any(axis=1)]

DF8 = DF8.apply(lambda x: x.fillna(x.median()),axis=0)

DF8['horsepower'] = DF8['horsepower'].astype('float64')

DF8.head()

DF8.info()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(DF8.corr(),annot=True)
```

```python
X = DF8.drop('mpg', axis=1)
y = DF8[['mpg']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

regression_model = LinearRegression()
regression_model.fit(X_train, y_train)

print("intercept = ",regression_model.intercept_)
```

```python
#Regression Coefficients
for idx, col_name in enumerate(X_train.columns):
    print("The Coefficient for {} is {}".format(col_name, regression_model.coef_[0][idx]))
```
